```python
import pygame
import random
import math
import sys
# initialize the pygame
pygame.init()
# create the screen
screen=pygame.display.set_mode((800,600))

#create the background
background=pygame.image.load("space.jpeg")
```

```python
# Title and Icon
pygame.display.set_caption("Space Invaders")
icon=pygame.image.load('ufo.png')
pygame.display.set_icon(icon)
```

```python
#Player 
playerImg=pygame.image.load('player.png')
playerX=370
playerY=480
playerX_change=0

#Enemy
enemyImg=[]
enemyX=[]
enemyY=[]
enemyX_change=[]
enemyY_change=[]
num_of_enemies=6

for i in range(num_of_enemies):
    enemyImg.append(pygame.image.load('alien.png'))
    enemyX.append(random.randint(0,700))
    enemyY.append(random.randint(20,400))
# enemy movement
    enemyX_change.append(0.1)
    enemyY_change.append(20)

# Bullet
#Ready - you cant see the bullet  on the screen
#Fire- The bullet is currently moving
bullet_state="ready"
bulletX=0
bulletY=480
bulletX_change=0
bulletY_change=.5
bulletIMG=pygame.image.load('bullet.png')
# score
score_val=0
font=pygame.font.Font('freesansbold.ttf',size=32)

textX=10
testY=10

#Game over text
game_over_font=pygame.font.Font('freesansbold.ttf',size=64)
def game_over_text():
        over_text=font.render("Game Over",True,(255,255,255))
        screen.blit(over_text,(250,250))

def show_score(x,y):
    score=game_over_font.render("Score: "+str(score_val),True,(255,255,255))
    screen.blit(score,(x,y))


def player(x,y):
    screen.blit(playerImg,(x,y))

def enemy(x,y,i):
    screen.blit(enemyImg[i],(x,y))


def fire_bullet(x,y):
    global bullet_state
    bullet_state="fire"
    screen.blit(bulletIMG,(x+25,y+10))

def isCollision(enemyX,enemyY,bulletX,bulletY):
    distance=math.sqrt((enemyX-bulletX)**2 +(enemyY-bulletY)**2)
    if distance<27:
        return True
    else:
        return False
```

```python
#game loop
running=True
while running:
    screen.fill((0,0,0))
    #background image
    screen.blit(background,(0,0))
    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            pygame.quit()
            sys.exit()
            running=False

        if event.type==pygame.KEYDOWN:
            if event.key==pygame.K_LEFT:
                playerX_change=-.2
            if event.key==pygame.K_RIGHT:
                playerX_change=.2
            if event.key==pygame.K_SPACE:
                if bullet_state=="ready":
                    bulletX=playerX
                    fire_bullet(bulletX,bulletY)

        if event.type==pygame.KEYUP:
            if event.key==pygame.K_LEFT or event.key==pygame.K_RIGHT:
                playerX_change=0

    playerX+=playerX_change

    if(playerX<=10):
        playerX=10
    if(playerX>=750):
        playerX=750

    for i in range(num_of_enemies):
        # Game Over
        if enemyY[i]>420:
            for j in range(num_of_enemies):
                enemyY[j]=2000
            game_over_text()
            break
        enemyX[i]+=enemyX_change[i]
        if(enemyX[i]<=10):
            enemyX_change[i]=0.1
            enemyY[i]+=enemyY_change[i]
        if(enemyX[i]>=750):
            enemyX_change[i]=-0.1
            enemyY[i]+=enemyY_change[i]

        # Collision
        collision=isCollision(enemyX[i],enemyY[i],bulletX,bulletY)
        if collision:
            bulletY=480
            bullet_state="ready"
            score_val+=1
            enemyX[i]=random.randint(0,700)
            enemyY[i]=random.randint(20,400)
        enemy(enemyX[i],enemyY[i],i)


    # Bullet Movement
    if bulletY <= 0:
        bulletY = 480
        bullet_state = "ready"

    if bullet_state=="fire":
        fire_bullet(bulletX,bulletY)
        bulletY-=bulletY_change

    player(playerX,playerY)
    show_score(textX,testY)
    pygame.display.update()
```
